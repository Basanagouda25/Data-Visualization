```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib Basics

```python
x = np.linspace(0,5,11) # linspace = linear space
x
y = x ** 2
y
```

```python
# linear graph
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

# sub plots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

# Object oriented way

```python
# to acess from a particular subplot
fig = plt.figure(figsize =(8,5))
axis1 = fig.add_axes([0.1,0.1,1,1]) # always give the values from 0-1
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```

# types of plots

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

# Saving plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```

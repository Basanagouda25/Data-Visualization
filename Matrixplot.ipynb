```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

# Heat Map

```python
tipscorr = tips[['total_bill','tip','size']]
```

```python
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

# Cluster Map

```python
sns.clustermap(tipscorr.corr(),annot=True)
```

# Pivot Table Heat Map

```python
pvtflight = flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvtflight
```

```python
sns.heatmap(pvtflight)
```

# regresion plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
# this is used in linear regiression
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```
